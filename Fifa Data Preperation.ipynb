```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
players_19=pd.read_csv("players_19.csv")
players_20=pd.read_csv("players_20.csv")
players_21=pd.read_csv("players_21.csv")
players_22=pd.read_csv("players_22.csv")
```

```python
players_19=players_19[["short_name", "player_positions", "overall", "potential", "age", "height_cm",
                      "weight_kg", "club_name", "league_name", "nationality_name", "preferred_foot",
                       "pace", "shooting", "passing", "dribbling", "defending"]]
players_20=players_20[["short_name", "player_positions", "overall", "potential", "age", "height_cm",
                      "weight_kg", "club_name", "league_name", "nationality_name", "preferred_foot",
                       "pace", "shooting", "passing", "dribbling", "defending"]]
players_21=players_21[["short_name", "player_positions", "overall", "potential", "age", "height_cm",
                      "weight_kg", "club_name", "league_name", "nationality_name", "preferred_foot",
                       "pace", "shooting", "passing", "dribbling", "defending"]]
players_22=players_22[["short_name", "player_positions", "overall", "potential", "age", "height_cm",
                      "weight_kg", "club_name", "league_name", "nationality_name", "preferred_foot",
                       "pace", "shooting", "passing", "dribbling", "defending"]]
```

```python
players_19=players_19[players_19["overall"]>80]
players_20=players_20[players_20["overall"]>80]
players_21=players_21[players_21["overall"]>80]
players_22=players_22[players_22["overall"]>80]

players_19["year"]=19
players_20["year"]=20
players_21["year"]=21
players_22["year"]=22

players=pd.concat([players_19, players_20, players_21, players_22], axis=0)
players=players.sort_values(by="overall", ascending=False)
players.head()
```
